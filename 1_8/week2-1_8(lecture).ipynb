## 형태소 단위 Tokenizer

```python
!pip install konlpy
```

```python
from konlpy.tag import Kkma

# Kkma 인스턴스 생성
kkma = Kkma()

# 테스트 문장
sentence = "한국어 분석을 시작합니다. 재미있겠어요!"

# 형태소 분석
print(kkma.morphs(sentence))

# 품사 태깅
print(kkma.pos(sentence))

# 문장 분리
print(kkma.sentences(sentence))
```

```python
from konlpy.tag import Komoran

# Komoran 인스턴스 생성
komoran = Komoran()

# 테스트 문장
sentence = "한국어 분석을 시작합니다. 재미있겠어요!"

# 형태소 분석
print(komoran.morphs(sentence))

# 품사 태깅
print(komoran.pos(sentence))
```

```python
from konlpy.tag import Okt

# Okt 인스턴스 생성
okt = Okt()

# 테스트 문장
sentence = "한국어 분석을 시작합니다. 재미있겠어요!"

# 형태소 분석
print(okt.morphs(sentence))

# 품사 태깅
print(okt.pos(sentence))

# 명사만 추출
print(okt.nouns(sentence))
```

```python
from konlpy.tag import Hannanum

# Hannanum 인스턴스 생성
hannanum = Hannanum()

# 테스트 문장
sentence = "한국어 분석을 시작합니다. 재미있겠어요!"

# 형태소 분석
print(hannanum.morphs(sentence))

# 품사 태깅
print(hannanum.pos(sentence))
```

## Subword 토크나이저

```python
!pip install transformers
```

```python
from transformers import BertTokenizer

# BERT 토크나이저 로드
tokenizer = BertTokenizer.from_pretrained('klue/bert-base')

# 텍스트 토큰화
text = "한국어 분석을 시작합니다. 재미있겠어요!"
encoded_input = tokenizer(text)

tokenize_output = tokenizer.tokenize(text)

print(encoded_input)
print(tokenize_output)
```

## 토큰화 실습 1
### 활용 데이터: sample.xlsx

### 실습 개요
이 실습에서는 주어진 xlsx 파일에 포함된 텍스트 데이터에 대해 Hannanum, Komoran 형태소 분석기를 사용하여 형태소 분석을 수행하고, 그 결과를 기존 데이터 프레임에 새로운 열로 추가하는 과제입니다. KoNLPy 라이브러리 Hannanum, Komoran 형태소 분석기를 활용하여 한국어 텍스트 데이터를 형태소 단위로 분리하고, 분리된 형태소를 각 행에 대응하는 새로운 열에 저장합니다.


### 실습 절차
1. 라이브러리 설치 및 임포트: 필요한 라이브러리 (konlpy)를 설치하고 임포트합니다.
2. xlsx 파일 로드: pd.read_excel()을 활용하여 엑셀파일을 로드합니다.
3. Hannanum, Komoran 인스턴스 생성: KoNLPy의 Hannanum, Komoran 클래스를 인스턴스화합니다.
4. 형태소 분석 수행: "원문"열에 대해 apply함수를 사용하여 형태소 분석기로 형태소 분석을 수행합니다. (morphs 함수 활용)
5. 각 행의 텍스트 데이터에 대한 형태소 분석 결과를 새로운 (hannanum, komoran)에 저장합니다.
6. 결과 확인 및 저장: 형태소 분석 결과가 추가된 DataFrame을 확인한 다음 morphs_results.csv파일로 저장합니다.

3. 토크나이저 로드: transformers를 활용해 토크나이저를 읽습니다.
4. 토크나이징: target_sentence에 대해 토크나이징을 수행합니다.
5. 토크나이징한 결과를 새로운 칼럼(tokenize_rst)에 저장합니다.
6. 결과 확인 및 저장: DataFrame을 확인한 다음 tokenize_results.csv파일로 저장합니다.
