## 토큰화 실습 1
### 활용 데이터: sample.xlsx

### 실습 개요
이 실습에서는 주어진 xlsx 파일에 포함된 텍스트 데이터에 대해 Hannanum, Komoran 형태소 분석기를 사용하여 형태소 분석을 수행하고, 그 결과를 기존 데이터 프레임에 새로운 열로 추가하는 과제입니다. KoNLPy 라이브러리 Hannanum, Komoran 형태소 분석기를 활용하여 한국어 텍스트 데이터를 형태소 단위로 분리하고, 분리된 형태소를 각 행에 대응하는 새로운 열에 저장합니다.


### 실습 절차
1. 라이브러리 설치 및 임포트: 필요한 라이브러리 (konlpy)를 설치하고 임포트합니다.
2. xlsx 파일 로드: pd.read_excel()을 활용하여 엑셀파일을 로드합니다.
3. Hannanum, Komoran 인스턴스 생성: KoNLPy의 Hannanum, Komoran 클래스를 인스턴스화합니다.
4. 형태소 분석 수행: "원문"열에 대해 apply함수를 사용하여 형태소 분석기로 형태소 분석을 수행합니다. (morphs 함수 활용)
5. 각 행의 텍스트 데이터에 대한 형태소 분석 결과를 새로운 (hannanum, komoran)에 저장합니다.
6. 결과 확인 및 저장: 형태소 분석 결과가 추가된 DataFrame을 확인한 다음 morphs_results.csv파일로 저장합니다.

```python
import pandas as pd
```

```python
df = pd.read_excel("/content/drive/MyDrive/NLP 실습/1 8/sample.xlsx")
```

```python
df.head(2)
```

```python
from konlpy.tag import Komoran, Hannanum

komoran = Komoran()
hannanum = Hannanum()
```

```python
df['komoran'] = df['원문'].apply(lambda x: komoran.morphs(x))
df['hannanum'] = df['원문'].apply(lambda x: hannanum.morphs(x))
```

```python
df.head(2)
```

```python
df.to_csv('morphs_results.csv', index=False)
```

## 토큰화 실습 2

### 활용 데이터: [HuggingFace 데이터](https://huggingface.co/datasets/ayymen/Pontoon-Translations/viewer/ko-en)

### 실습 개요
이 실습에서는 HuggingFace의 데이터 셋을 활용하여 [KoBERT](https://huggingface.co/skt/kobert-base-v1)를 활용하여 토크나이징을 수행하고 그 결과를 기존 데이터 프레임에 새로운 열로 추가하는 과제입니다.

### 실습 절차
1. 라이브러리 설치 및 임포트: 필요한 라이브러리 (transformers, datasets, sentencepiece)를 설치하고 임포트 합니다.
2. HuggingFace의 데이터 로드: datasets를 활용해 데이터를 읽고 DataFrame으로 변환합니다. (dataset = load_dataset("ayymen/Pontoon-Translations", 'ko-en') 코드를 활용합니다.)
3. 데이터 프레임 중, 1000개의 데이터만 활용합니다. (df = df[:1000] 코드를 활용합니다.)
3. 토크나이저 로드: transformers를 활용해 토크나이저를 읽습니다.
4. 토크나이징: target_sentence에 대해 토크나이징을 수행합니다.
5. 토크나이징한 결과를 새로운 칼럼(tokenize_rst)에 저장합니다.
6. 결과 확인 및 저장: DataFrame을 확인한 다음 tokenize_results.csv파일로 저장합니다.

```python
!pip install transformers datasets sentencepiece
```

```python
from datasets import load_dataset
import pandas as pd
```

```python
dataset = load_dataset("ayymen/Pontoon-Translations", 'ko-en', split='train')
```

```python
dataset
```

```python
df = pd.DataFrame()

df['source_sentence'] = dataset['source_sentence']
df['target_sentence'] = dataset['target_sentence']
```

```python
df = df[:1000]
```

```python
from transformers import AutoTokenizer, AutoModel
```

```python
tokenizer = AutoTokenizer.from_pretrained("skt/kobert-base-v1")
```

```python
df['tokenize_rst'] = df['target_sentence'].apply(lambda x: tokenizer.tokenize(x))
```

```python
df.head(2)
```

```python
df.to_csv('tokenize_results.csv', index=False)
```
