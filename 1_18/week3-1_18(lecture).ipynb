```python
import konlpy
from nltk.translate.meteor_score import meteor_score
from konlpy.tag import Okt

# 한국어 문장을 토크나이즈하는 함수
def tokenize_korean(sent):
    okt = Okt()
    return okt.morphs(sent)
```

```python
# 참조 문장과 번역된 문장
reference = "전 학교에 갑니다."
candidate = "난 학교로 갑니다."
```

```python
# 토크나이즈 수행
tokenized_ref = tokenize_korean(reference)
tokenized_can = tokenize_korean(candidate)
```

```python
print(tokenized_ref)
print(tokenized_can)
```

```python
print(tokenized_ref)
print(tokenized_can)
```

```python
# METEOR 점수 계산
score = meteor_score([tokenized_ref], tokenized_can)
print(score)
```

```python
# 필요한 라이브러리와 모델 다운로드
from laserembeddings import Laser
import numpy as np

# LASER 모델 초기화
laser = Laser()
```

```python
# 참조 문장과 번역된 문장
reference_sentences = "저는 학교에 갑니다"
translated_sentences = "I go to school"
```

```python
# 문장 임베딩 생성
reference_embeddings = laser.embed_sentences([reference_sentences], lang='ko')
translated_embeddings = laser.embed_sentences([translated_sentences], lang='en')
```

```python
reference_embeddings
```

```python
translated_embeddings
```

```python
# 코사인 유사도 계산
def cosine_similarity(v1, v2):
    n1 = np.linalg.norm(v1)
    n2 = np.linalg.norm(v2)
    return np.dot(v1, v2.T).squeeze() / n1 / n2
```

```python
print(f"Cosine similarity (LASER score): {cosine_similarity(reference_embeddings, translated_embeddings)}")
```

# chrF

```python
chrf = load_metric('chrf')
```

```python
chrf
```

```python
reference = "저는 학교에 갑니다"
candidate = "저는 학교에 가요"
```

```python
score = chrf.compute(predictions=[candidate], references=[[reference]]) # default(official) n-gram: 6 # original n-gram: 4(Popovic 2015)
print(score)
# n_char_order (int) – A character n-gram order. If n_char_order=6, the metrics refers to the official chrF/chrF++.
# n_word_order (int) – A word n-gram order. If n_word_order=2, the metric refers to the official chrF++. If n_word_order=0, the metric is equivalent to the original ChrF.
# beta (float) – parameter determining an importance of recall w.r.t. precision. If beta=1, their importance is equal.
```
