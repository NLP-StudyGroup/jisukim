```python
from nltk.translate.bleu_score import sentence_bleu, corpus_bleu
from konlpy.tag import Okt

# 토큰화
okt = Okt()

# 원문과 번역문
reference_text = "저는 학교에 갑니다"
candidate_text = "나는 학교에 갑니다"

# Okt 토크나이저를 사용하여 토큰화
reference_tokens = okt.morphs(reference_text)
candidate_tokens = okt.morphs(candidate_text)

# BLEU 점수 계산 - ver1
bleu_score = sentence_bleu([reference_tokens], candidate_tokens)
print(bleu_score)

# BLEU 점수 계산 - ver2
bleu_score = corpus_bleu([[reference_tokens]], [candidate_tokens])
print(bleu_score)
```

```python
# sacrebleu는 하나의 툴로써, Moses tokenizer를 활용해 토큰화한 결과를 기반으로 bleu 스코어를 계산합니다.
# 하지만, Moses tokenizer는 공백을 기반으로 단어를 분리하기 때문에, 한국어로 활용하기엔 부족합니다.
# 따라서, 한국어 번역모델을 평가할 때는 sacrebleu 패키지를 활용하여 한국어에 맞는 토크나이저를 활용해 bleu 스코어를 계산해야 합니다.
# 이때, bleu 패키지와 sacrebleu 패키지에서 동일한 토크나이저를 활용했을 경우, 계산 결과는 동일하게 나옵니다.

import sacrebleu

# 토큰화
okt = Okt()

# 참조 문장과 후보 문장
references = ["저는 학교에 갑니다"]
candidates = ["나는 학교에 갑니다"]

# 토크나이징 수행
tokenized_references = [okt.morphs(ref) for ref in references]
tokenized_candidates = [okt.morphs(cand) for cand in candidates]

# 토크나이징된 문장을 다시 문자열로 변환
detokenized_references = [' '.join(ref) for ref in tokenized_references]
detokenized_candidates = [' '.join(cand) for cand in tokenized_candidates]

# BLEU 점수 계산
bleu = sacrebleu.corpus_bleu(detokenized_candidates, [detokenized_references])
print(f"BLEU score: {bleu.score}")
```

```python
# 아래와 같이 sacrebleu 패키지 내에 내장되어있는 Moses tokenizer를 활용해 bleu 값을 출력할 경우 값이 다르게 출력됩니다.

from sacrebleu import corpus_bleu

# 참조 문장과 후보 문장
references = ["저는 학교에 갑니다"]
candidates = ["나는 학교에 갑니다"]

# SacreBLEU 점수 계산
bleu = sacrebleu.corpus_bleu(candidates, [references])

print(f"BLEU score: {bleu.score}")
```
