## 데이터 정제

### 활용 데이터: hw_sample.json

### 실습 개요
이 실습에서는 Google Colab에 json 파일을 읽은 다음, Json 데이터를 정제하고 Pandas 라이브러리를 사용하여 데이터 프레임으로 변환한 다음, 특정 컬럼의 데이터를 분석합니다. 'src_sentence'과 'tgt_sentence' 라는 두 개의 컬럼에 대해 각각 띄어쓰기로 분할하여 단어의 수를 세고, 이 정보를 새로운 컬럼에 저장합니다.
  
  
### 실습 절차
1. json 파일 읽기: Google Colab에서 json 파일을 읽습니다.
2. json 파일 정제: json 파일에서 paragraph > sentences 의 데이터를 가져옵니다.
3. 데이터 읽기: Pandas를 사용하여 json파일을 DataFrame으로 변환합니다.
4. 데이터 처리:
- 'src_sentence'과 'tgt_sentence' 컬럼의 데이터를 띄어쓰기 단위로 분할합니다.
- 각 행에 대해 영어와 한국어 단어의 개수를 세고, 이를 새로운 컬럼에 저장합니다.
4. tgt_sentence 칼럼에 대해 길이가 긴 순서대로 DataFrame을 재 정렬합니다. 이때, tgt_sentence의 길이가 같을 경우, src_sentence가 긴 순서대로 재 정렬합니다.
5. 해당 데이터 파일을 csv로 저장합니다.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import json
import pandas as pd
```

```python
with open("/content/drive/MyDrive/NLP정답/1-4/hw_sample.json") as f:
  data = json.load(f)
```

```python
df = pd.DataFrame(data['paragraph'][0]['sentences'])
```

```python
df.head(2)
```

```python
df['en_count'] = df['tgt_sentence'].apply(lambda x: len(x.split()))
df['ko_count'] = df['src_sentence'].apply(lambda x: len(x.split()))
```

```python
df = df.sort_values(by=['en_count','ko_count'], ascending=[False, False])
```

```python
df.head(3)
```

```python
df.to_csv('homework.csv', index=False)
```
